# **Lane Detection using OpenCV**

### **0. Do the necessary imports**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### **1 a) Read the image**

```python
img = cv2.imread("images/test.png")
```

### **1 b) Make a copy of the original image**

```python
lane_img = np.copy(img)
```

### **2. Convert the image to a grayscale image**

```python
gray = cv2.cvtColor(lane_img, cv2.COLOR_RGB2GRAY)
cv2.imshow("Gray", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(gray)
plt.show()
```

### **3. Apply Gaussian Filtering to smoothen out the image and reduce noise**

```python
g_blur = cv2.GaussianBlur(gray, (5,5), 0)
cv2.imshow("Gaussian Blur", g_blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(g_blur)
plt.show()
```

### **4. Apply the canny method to detect the edges in the image**

```python
canny = cv2.Canny(g_blur, 50, 150)
cv2.imshow("Canny Edge Detection", canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(canny)
plt.show()
```

### **5. Specify the Region on Interest (ROI)**

#### **a) Create a Triangle Shaped Polygon**  

```python
height = canny.shape[0]
polygons = np.array([
    [(200, height), (1100, height), (550, 250)]
])
```

#### **b) Create a black background with same dimension as the original image**

```python
mask = np.zeros_like(canny)
```

#### **c) Use the fillPoly method to create a mask with the enclosed polygon**

```python
cv2.fillPoly(mask, polygons, (255,255,255))
plt.imshow(mask)
plt.show()
cv2.imshow("Mask", mask)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### **6. Use the Bitwise_and operation to mask the canny image and get only the lanes**

```python
masked_img = cv2.bitwise_and(canny, mask)
plt.imshow(masked_img)
plt.show()
cv2.imshow("Masked Canny Image", masked_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
